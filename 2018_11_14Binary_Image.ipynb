# Try all Threshold
Converting the image to a binary image using various methods. 

```python
import matplotlib.pyplot as plt

import skimage.io
from skimage.filters import try_all_threshold
from skimage.color import rgb2gray

imname = 'HBBS_vid_1_overlay_frame_1.tif'
img = skimage.io.imread(imname)

img = img[:, :, 1]

img_gray = rgb2gray(img)

fig, ax = try_all_threshold(img_gray, figsize=(6, 8), verbose=False)
plt.show()

fig.savefig('PNN_Try_All_Threshold')
```

# Threshold Mean
1. Trying one method of converting the image to a binary image.
2. Viewing the image.

```python
import matplotlib.pyplot as plt

import skimage.io
from skimage.filters import threshold_mean
from skimage.color import rgb2gray

imname = 'Slice_4_40X_Top_of_Brain.tif'
img = skimage.io.imread(imname)

img_gray = rgb2gray(img)

thresh = threshold_mean(img_gray)
binary = img_gray > thresh
```

```python
import skimage.io
from skimage.viewer import ImageViewer
veiwer = ImageViewer(binary)
veiwer.show()
```
