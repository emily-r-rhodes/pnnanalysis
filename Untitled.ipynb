```python
import matplotlib.pyplot as plt

import skimage.io
from skimage.filters import try_all_threshold
from skimage.filters import threshold_isodata
from skimage.color import rgb2gray

imname = 'Ryan_Logo.tif'
img = skimage.io.imread(imname)

img_gray = rgb2gray(img)

fig, ax = try_all_threshold(img_gray, figsize=(6, 8), verbose=False)
plt.show()

#fig.savefig('PNN_Try_All_Threshold')
```

```python
fig = threshold_isodata(img_gray)
plt.show()
```

```python
import skimage.io
from skimage.viewer import ImageViewer
veiwer = ImageViewer(threshold_isodata(img_gray))
veiwer.show()
```
