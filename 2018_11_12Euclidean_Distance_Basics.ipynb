# Euclidean Distance

Two points:
 (xa, ya, za)
 (xb, yb, zb)
 
Calculate the distance: distance = sqrt((xa-xb)^2 + (ya-yb)^2 + (za-zb)^2)

```python
xa = 1
ya = 1
za = 1
xb = 0
yb = 0
zb = 0

import numpy as np
a = np.array((xa, ya, za))
b = np.array((xb, yb, zb))

dist = np.linalg.norm(a-b)
```

# Euclidean Distance in Images

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('Origional.Practice-Euclidean.Distance.png')
```

```python
import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure(figsize =(4,4))
a = fig.add_subplot(1,1,1)
imgplot = plt.imshow(img)
a.set_title('Origional', fontsize = 20)
a.set_axis_off()
 
fig.savefig('Origional_From_Chad.png')
```
