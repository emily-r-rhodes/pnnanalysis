# Image to Binary

```python
import matplotlib.pyplot as plt

import skimage.io
from skimage.filters import threshold_mean
from skimage.color import rgb2gray

imname = 'Slice_4_40X_Top_of_Brain.tif'
img = skimage.io.imread(imname)

img_gray = img[:, :, 0]

thresh = threshold_mean(img_gray)
binary = img_gray > thresh

import matplotlib.pyplot as plt

plt.imshow(binary, cmap='gray')
plt.show()
```

```python
from skimage.morphology import remove_small_objects
```

```python
clim = remove_small_objects(binary, min_size=100)
plt.imshow(clim)
plt.show()
```

```python
from scipy.ndimage.morphology import distance_transform_edt as Euclidean
euim = Euclidean(1-clim>0)

plt.imshow(euim)
plt.show()
```

```python
from scipy import ndimage as ndi
distance = ndi.distance_transform_edt(1-clim)

plt.imshow(distance)
plt.show()
```

```python
lbim = skimage.measure.label(clim) #clean image that is labeled with each object being a different number

plt.imshow(lbim)
plt.show()
```

```python
prop = skimage.measure.regionprops(lbim)

for prop in lbim:
    print(prop, lbim[prop])
```
